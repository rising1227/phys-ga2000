```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussian(sigma,dev,x):
    a = (1/(sigma*np.sqrt(2*np.pi))) * np.e**(-0.5*(x-dev)**2/sigma**2)
    return a
```

```python
x = np.linspace(-10,10,100)
```

```python
y = gaussian(3,0,x)
```

```python
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Gaussian Function with zero mean and standard derivation of 3")
plt.plot(x,y)
```
